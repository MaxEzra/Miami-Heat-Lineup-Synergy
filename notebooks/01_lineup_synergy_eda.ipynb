```python
import pandas as pd
from pathlib import Path

project_root = Path.cwd().parent
raw_path = project_root / "data" / "raw" / "heat_lineups_5man_2025-26_Regular_Season.csv"

df = pd.read_csv(raw_path)
df.head()

columns = [ 
    "GROUP_NAME",
    "GP",
    "MIN",
    "OFF_RATING",
    "DEF_RATING",
    "NET_RATING",
]

lineups = df[columns].copy()
lineups.head()
```

```python
columns = [ 
    "GROUP_NAME",
    "GP",
    "MIN",
    "OFF_RATING", 
    "DEF_RATING",
    "NET_RATING"
]

lineups = df[columns].copy()
lineups.head()
```

```python
MIN_LIMIT = 25

filtered = lineups[lineups["MIN"] >= MIN_LIMIT].copy()
filtered.shape
```

```python
filtered_sorted = (
    filtered.sort_values("NET_RATING", ascending=False)
    .reset_index(drop=True)
)

filtered_sorted.head(10)
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(12, 6))

plt.barh(
    top_lineups["GROUP_NAME"][::-1],
    top_lineups["NET_RATING"][::-1],
    color='#98002E' # Fire Red
)

plt.xlabel("Net Rating", fontsize=12)
plt.ylabel("Lineup", fontsize=12)
plt.title(f"Top {top_n} Miami Heat 5-Man Lineups by Net Rating", fontsize=14)

plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()
```

```python
def split_players(group_name): 
    return [p.strip() for p in group_name.split('-')]

filtered_sorted["PLAYERS"] = filtered_sorted["GROUP_NAME"].apply(split_players)
filtered_sorted[["GROUP_NAME", "PLAYERS"]].head()
```

```python
filtered_sorted["PLAYERS"].head(20).tolist()
```

```python
player_attributes = {
    "B. Adebayo":  {"shooting": False, "defense": True,  "creator": False, "position": "C"},
    "D. Smith":    {"shooting": True,  "defense": True,  "creator": False, "position": "G"},
    "N. Jović":    {"shooting": True,  "defense": True,  "creator": False, "position": "F"},
    "J. Jaquez Jr.": {"shooting": True, "defense": True, "creator": True, "position": "F"},
    "S. Fontecchio": {"shooting": True, "defense": False, "creator": False, "position": "F"},
    
    "A. Wiggins":  {"shooting": False, "defense": True,  "creator": False, "position": "W"},  
    "N. Powell":   {"shooting": True,  "defense": False, "creator": False, "position": "G"},  
    "D. Mitchell": {"shooting": True,  "defense": True,  "creator": True, "position": "G"},   
    "P. Larsson":  {"shooting": True,  "defense": False, "creator": False, "position": "F"},
    "K. Ware":     {"shooting": False, "defense": True,  "creator": False, "position": "C"},
    "T. Herro":    {"shooting": True,  "defense": False, "creator": True, "position": "G"},
}
```

```python
def is_shooting_heavy(players):
    shooters = sum(player_attributes[p]["shooting"] for p in players if p in player_attributes)
    return shooters >= 3

def is_defense_heavy(players):
    defenders = sum(player_attributes[p]["defense"] for p in players if p in player_attributes)
    return defenders >= 3

def is_switchable(players):
    wings = sum(player_attributes[p]["position"] in ["F", "W"] for p in players if p in player_attributes)
    return wings >= 3

def has_primary_creator(players):
    creators = sum(player_attributes[p]["creator"] for p in players if p in player_attributes)
    return creators >= 1

def spacing_around_bam(players):
    # Lineups where Bam + at least 3 shooters are on the floor
    if "B. Adebayo" not in players:
        return False
    shooters = sum(player_attributes[p]["shooting"] for p in players if p in player_attributes)
    return shooters >= 3
```

```python
def generate_tags(players):
    tags = []
    
    if is_shooting_heavy(players):
        tags.append("Shooting-Heavy")
    if is_defense_heavy(players):
        tags.append("Defense-Heavy")
    if is_switchable(players):
        tags.append("Switchable")
    if has_primary_creator(players):
        tags.append("Creator Present")
    if spacing_around_bam(players):
        tags.append("Bam + Spacing")
    
    if len(tags) == 0:
        tags.append("Balanced")
        
    return tags


filtered_sorted["TAGS"] = filtered_sorted["PLAYERS"].apply(generate_tags)
filtered_sorted[["GROUP_NAME", "NET_RATING", "TAGS"]].head(10)
```

```python
def primary_tag(tag_list): 
    return tag_list[0] if len(tag_list) > 0 else "Other"

filtered_sorted["PRIMARY_TAG"] = filtered_sorted["TAGS"].apply(primary_tag)
filtered_sorted[["GROUP_NAME", "PRIMARY_TAG"]].head()

color_map = {
    "Shooting-Heavy": "#E03A3E",   # Heat red
    "Defense-Heavy": "#000000",    # Black
    "Switchable": "#F9A01B",       # Heat gold
    "Creator Present": "#21A6F3",  # Blue for playmaking
    "Bam + Spacing": "#00A86B",    # Green
    "Balanced": "#7F7F7F",         # Gray
}
```

```python
full_name_map = {
    "B. Adebayo": "Bam Adebayo",
    "D. Smith": "D. Smith",
    "N. Jović": "Nikola Jović",
    "J. Jaquez Jr.": "Jaime Jaquez Jr.",
    "S. Fontecchio": "Simone Fontecchio",
    "A. Wiggins": "Andrew Wiggins",
    "N. Powell": "Norman Powell",
    "D. Mitchell": "Davion Mitchell",
    "P. Larsson": "P. Larsson",
    "K. Ware": "Kel’el Ware",
    "T. Herro": "Tyler Herro"
}

def format_lineup_label(players):
    formatted = []
    for p in players:
        full = full_name_map.get(p, p)
        last = full.split()[-1]  # Only show last name for clarity
        formatted.append(last)
    return " | ".join(formatted)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 9))

# Scatter plot by synergy tag
for tag, subset in filtered_sorted.groupby("PRIMARY_TAG"):
    plt.scatter(
        subset["OFF_RATING"],
        subset["DEF_RATING"],
        s=subset["MIN"] * 5,
        c=color_map.get(tag, "#7F7F7F"),
        alpha=0.7,
        label=tag
    )

# Add lineup labels to each point using last names
for _, row in filtered_sorted.iterrows():
    label = format_lineup_label(row["PLAYERS"])
    plt.text(
        row["OFF_RATING"] + 0.1,
        row["DEF_RATING"] + 0.1,
        label,
        fontsize=7.5,
        alpha=0.85
    )

# Quadrant medians
off_median = filtered_sorted["OFF_RATING"].median()
def_median = filtered_sorted["DEF_RATING"].median()

plt.axvline(off_median, color="gray", linestyle="--", alpha=0.7)
plt.axhline(def_median, color="gray", linestyle="--", alpha=0.7)

# Quadrant labels
plt.text(off_median + 0.5, def_median - 0.5, "Best Quadrant", fontsize=12, color="green")
plt.text(off_median - 5, def_median + 1, "Defensive Lineups", fontsize=10, color="blue")
plt.text(off_median + 0.5, def_median + 1, "Offensive Lineups", fontsize=10, color="orange")
plt.text(off_median - 5, def_median - 0.5, "Worst Quadrant", fontsize=12, color="red")

# Titles and labels
plt.title("Miami Heat Lineup Synergy Scatterplot", fontsize=16)
plt.xlabel("Offensive Rating (Higher is Better)", fontsize=12)
plt.ylabel("Defensive Rating (Lower is Better)", fontsize=12)
plt.gca().invert_yaxis()  # Lower defensive rating is better

plt.legend(title="Synergy Tags")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
